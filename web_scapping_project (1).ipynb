#lunch data

```python
import re
import requests
import numpy as np
from bs4 import BeautifulSoup
lunch_data=[]
rate=[]
location_data=[]
food_type=[]
food_price=[]
offer_data=[]
for i in range(1,21):
  url=f'https://www.eazydiner.com/restaurants?date=2025-06-25&meal_period=lunch&location=hyderabad&page={i}'
  page=requests.get(url)
  lun=BeautifulSoup(page.text)
  lunc=lun.find_all("a",class_='ellipsis listing_res_name__uVIN8')
  for i in lunc:
   l=i.text
   lunch_data.append(l)

  rating=lun.find_all("div",class_="flex align-v-center flex-between")

  for i in rating:
   txt=i.text
   regx=re.findall(r'\d.\d',txt)
   if regx:
     rate.append(regx[0])


  loc=lun.find_all("div",class_="font-12 ellipsis flex listing_res_address__LuVOm")

  for tag in loc:
    tag=tag.text
    loo=re.findall(r'(?:^|,\s)\s*([^,]+?)\s*(?:, Hyderabad|$)',tag)
    location_data.append(loo)

  foodtype=lun.find_all('div',class_='flex align-v-center ellipsis listing_multi_cuisines__zLLXK')

  for i in foodtype:
   f=i.text
   food_type.append(f)


  foodprice=lun.find_all('div',class_="flex align-v-center")

  for item in foodprice:
     pri = item.text.strip()
     price = re.findall(r'\₹\s(\d+)', pri)
     if price:
         food_price.append(price[0])
  offer=lun.find_all('div',class_="ellipsis listing_offer_text__LXKw1")

  for item in offer:
     it = item.text
     offer_data.append(it)
print(lunch_data)
print(rate)
print(location_data)
print(food_type)
print(food_price)
print(offer_data)
```

```python
import re
from bs4 import BeautifulSoup
import requests
import numpy as np
hotel=[]
rate=[]
loc_restaurants=[]
food_type=[]
price_food=[]
offer_data=[]
for i in range(1,21):
  url=f'https://www.eazydiner.com/restaurants?bookable=1&accepts_eazypay_wallet=1&rest_tags%5B%5D=fast_food&location=hyderabad&page={i}'
  page=requests.get(url)
  fastfood=BeautifulSoup(page.text)
# hotel name
  fast=fastfood.find_all("a",class_="ellipsis listing_res_name__uVIN8")
  for i in fast:
    var=i.text

    if var:
      hotel.append(var)
    else:
      hotel.append(np.nan)
#rate
  rating=fastfood.find_all("div",class_="flex align-v-center flex-between")

  for i in rating:
   txt=i.text
   regx=re.findall(r'\d.\d',txt)
   if regx:
     rate.append(regx[0])

# location

  loc_re=fastfood.find_all("div",class_="font-12 ellipsis flex listing_res_address__LuVOm")

  for tag in loc_re:
    tag=tag.text
    loo=re.findall(r'(?:^|,\s)\s*([^,]+?)\s*(?:, Hyderabad|$)',tag)

    if loo:
      loc_restaurants.append(loo)
    else:
      loc_restaurants.append(np.nan)

# foodtype
  foodtype=fastfood.find_all('div',class_='flex align-v-center ellipsis listing_multi_cuisines__zLLXK')
  for i in foodtype:
    f=i.text
    if f:
      food_type.append(f)
    else:
      food_type.append(np.nan)

# price
  foodprice=fastfood.find_all('div',class_="flex align-v-center")
  for item in foodprice:
    pri = item.text.strip()

    price = re.findall(r'\₹\s(\d+)', pri)

    if price:
        price_food.append(price[0])

# offer
  offer=fastfood.find_all('div',class_="ellipsis listing_offer_text__LXKw1")
  for item in offer:
     it = item.text

     offer_data.append(it)
```

```python
import re
df = eazydiner_data['offer_data']

def consolidate_offers(offer):

    percentages = re.findall(r'(\d+)%', offer)

    if percentages:
        total = sum(int(p) for p in percentages)
        return f"{total}% Off"

    return offer


eazydiner_data['offer_data'] = eazydiner_data['offer_data'].apply(consolidate_offers)
```
